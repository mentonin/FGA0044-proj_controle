### Compensadores de atraso e avanço

Compensador de avanço de fase:
$$
K_c \frac{s + \frac{1}{T}}{s + \frac{1}{\alpha T}} \qquad 0 < \alpha < 1
$$

Compensador de atraso de fase:
$$
K_c \frac{s + \frac{1}{T}}{s + \frac{1}{\beta T}} \qquad \beta > 1
$$

```python
import control
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = control.TransferFunction([1], [1, 3, 2, 0])
G_ = control.feedback(G)
```

```python
K = 1
Gc = control.TransferFunction([1, .05], [1, .005])
Gc_ = control.feedback(G*Gc)
Gc2 = control.TransferFunction([1, .0005], [1, .00005])
Gc2_ = control.feedback(G*Gc2)
```

```python
t = np.linspace(0, 70, 5000)
res = control.step_response(G_, T=t)
res_c = control.step_response(Gc_, T=t)
res_c2 = control.step_response(Gc2_, T=t)
# plt.plot(res[0], res[1])
plt.plot(res_c[0], res_c[1])
plt.plot(res_c2[0], res_c2[1])
plt.show()
```
